```python
import numpy as np
import matplotlib.pyplot as plt
from utils.units import *
```

```python
lmbda = 1529*nm
w0 = 1.5*mm
f = 100*mm
w1 = lmbda*f/(w0*np.pi)
print(f"focused waist: {w1/um} um")
zR = np.pi*w1**2/lmbda
print(f"Rayleigh: {zR/um} um")
```

```python
z = np.linspace(-100*um,100*um,100)
wz = w1*np.sqrt(1+(z/zR)**2)
plt.plot(z/um, wz/w1)
plt.xlabel("z (um)")
plt.ylabel("w/w_focus")
plt.title("w(z)/w0 vs defocus, f=100 mm lens, 1529 nm w0=1.5mm input")
```
